```python
import pandas as pd

df = pd.read_csv('../CustomQuestions.csv')
df
```

```python
final = []

for _, row in df.iterrows():
    idx = row['Image Indices']
    question = row['Question']
    options = question.split('? ')[-1]
    question = question.split('a)')[0]

    assert question
    assert options 
    # print(row)
    # print(question)
    # print(options)


    new_entry = {
        'Index': idx,
        'Question': question,
        'Options': options, 
        'Correct Answer': ''
    }

    final.append(new_entry)
    new_entry2 = {
        'Index': idx+1,
        'Question': question,
        'Options': options, 
        'Correct Answer': ''
    }
    final.append(new_entry2)
```

```python
final_qs = pd.DataFrame(final)
final_qs.to_csv('CustomQuestions.csv', index=False)
```
